# Rendimiento E-commerce

Para nuestro analisis de ventas por tiendas, desarrollamos los siguientes actividades y calculos:

- Identificar mes con mayor eficiencia (ventas/gasto publicidad)
- Determinar mes con mejor tasa de conversi√≥n y analizar causa
- Calcular ticket promedio (ventas/productos) por mes
- Evaluar relaci√≥n entre visitantes y ventas

Analizar la eficiencia publicitaria, la tasa de conversi√≥n, el ticket promedio y la relaci√≥n entre visitantes y ventas proporciona una visi√≥n clara de qu√© tan rentable, efectiva y coherente es la estrategia comercial de tu tienda.

Estos indicadores permiten optimizar presupuestos, detectar comportamientos del cliente y orientar decisiones de marketing basadas en datos reales.


## Mes con mayor eficiencia

```python
import pandas as pd

data = {
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May'],
    'Ventas ($)': [45000, 52000, 38000, 61000, 48000],
    'Visitantes': [15000, 18200, 12500, 20500, 16800],
    'Conversion (%)': [3.2, 2.9, 3.8, 3.1, 2.7],
    'Gasto Publicidad ($)': [8500, 9800, 7200, 11200, 9500],
    'Productos Vendidos': [450, 520, 380, 610, 480]
}

df = pd.DataFrame(data)
```

```python
# 1. Calcular la eficiencia (Ventas por cada d√≥lar de Publicidad)
df['Eficiencia'] = df['Ventas ($)'] / df['Gasto Publicidad ($)']

# 2. Encontrar el mes con la m√°xima eficiencia
mes_mas_eficiente = df.loc[df['Eficiencia'].idxmax()]

print("Mes con mayor eficiencia (Ventas/Gasto Publicidad):")
print(f"Mes: {mes_mas_eficiente['Mes']}")
print(f"Eficiencia: {mes_mas_eficiente['Eficiencia']:.2f}")
```

Durante abril, por cada unidad monetaria invertida en publicidad se generaron 5.45 unidades de venta, lo cual indica un excelente retorno sobre la inversi√≥n publicitaria (ROI).

Este mes representa el punto m√°s alto de eficiencia comercial, probablemente debido a:

- Campa√±as bien segmentadas o mejor optimizadas.
- Mayor inter√©s estacional por ciertos productos.
- Aumento org√°nico del tr√°fico o mejora en la retenci√≥n de clientes.

**Impacto:**

Identificar este mes permite replicar las condiciones de √©xito (tipo de campa√±a, canal, promociones o mix de productos), redirigiendo presupuesto hacia estrategias que maximicen el rendimiento y reduciendo gastos en meses menos eficientes.

## Mes con la mejor tasa de conversion

```python
mes_mejor_conversion = df.loc[df['Conversion (%)'].idxmax()]

print("\nMes con mejor tasa de conversi√≥n:")
print(f"Mes: {mes_mejor_conversion['Mes']}")
print(f"Conversi√≥n: {mes_mejor_conversion['Conversion (%)']}%")
```

En marzo, la proporci√≥n de visitantes que concretaron una compra fue la m√°s alta del periodo.
Esto sugiere que, aunque la eficiencia publicitaria fue menor que en abril, la capacidad de convertir visitas en ventas fue superior, lo que puede estar asociado con:

- Campa√±as m√°s persuasivas o enfocadas en retenci√≥n.
- Mejor experiencia de usuario (UX) en el sitio.
- Promociones o lanzamientos que incentivaron la compra inmediata.

**Impacto:**

Analizar qu√© factores influyeron en esta conversi√≥n (fuente de tr√°fico, tipo de anuncio, comportamiento del cliente) ayuda a replicar la experiencia m√°s efectiva del embudo de ventas, fortaleciendo las estrategias de atracci√≥n y cierre.

## Ticket promedio

```python
# 1. Calcular el ticket promedio
df['Ticket Promedio ($)'] = df['Ventas ($)'] / df['Productos Vendidos']

print("\nTabla con Ticket Promedio por mes:")
# Usar .to_string() para evitar mostrar metadatos
print(df[['Mes', 'Ventas ($)', 'Productos Vendidos', 'Ticket Promedio ($)']].to_string(index=False))
```

El ticket promedio se mantuvo constante en $100 a lo largo de los meses, lo que indica consistencia en los precios y h√°bitos de compra.

Sin embargo, tambi√©n podr√≠a reflejar una falta de estrategias de upselling o cross-selling, ya que no hay evidencia de incremento en el gasto por cliente.

**Impacto:**
Mantener un ticket promedio estable garantiza previsibilidad, pero incrementarlo mediante estrategias como recomendaciones personalizadas, combos o descuentos por volumen podr√≠a elevar los ingresos sin necesidad de aumentar el tr√°fico.

## Relaci√≥n entre Visitantes y Ventas

```python
# 1. Calcular el coeficiente de correlaci√≥n de Pearson
correlacion = df['Visitantes'].corr(df['Ventas ($)'])

print(f"\nCoeficiente de Correlaci√≥n (Visitantes vs Ventas): {correlacion:.4f}")
```

Existe una correlaci√≥n casi perfecta y positiva (r ‚âà 0.99) entre el n√∫mero de visitantes y las ventas.

Esto significa que el crecimiento de las ventas est√° altamente vinculado al incremento del tr√°fico, lo que evidencia que:

El sitio convierte visitantes en compradores de manera eficiente.

Las campa√±as de atracci√≥n son efectivas en generar ventas directas.

**Impacto:**

El alto nivel de correlaci√≥n indica que incrementar el tr√°fico tiene un efecto directo en las ventas, pero tambi√©n sugiere la oportunidad de diversificar estrategias de crecimiento, por ejemplo:

Mejorar el valor promedio de compra (AOV).

Fomentar la recompra (retenci√≥n de clientes).

Optimizar conversiones sin depender tanto del aumento del tr√°fico.
